```python
import sys, os, subprocess
sys.path.append(os.path.abspath(os.path.join('..')))

subprocess.run("make -C ../libc clean && make -C ../libc", shell=True)

from engine.interop.loader import Loader
Loader.loadLibrary(
    lib_name="libc",
    lib_folder="../libc",
    build_folder="../libc/build",
    specs_folder="../libc/specs",
    dependencies_bin_folder=r"C:\msys64\mingw64\bin",
    seed=None
)

from engine.interop.rbf import RBF

X = [0.0, 1.0, 2.0, 3.0]
Y = [0.0, 2.0, 4.0, 6.0]

rbf = RBF(input_dim=1, num_centers=4)
rbf.train(X, Y, data_size=len(X), alpha=0.01, epochs=2000, is_classification=False)

for x, y in zip(X, Y):
    pred = rbf.predict([x], is_classification=False)
    print(f"x={x} -> prédit={pred:.2f} (attendu={y})")
```
