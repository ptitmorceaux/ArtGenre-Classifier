```python
# Import

import sys
import os
import numpy as np

sys.path.append(os.path.abspath(os.path.join('..')))
from api.wrappers.mathlib import MathLib

try:
    mathlib = MathLib(lib_folder="../libc", build_folder="build", specs_folder="specs", basename="mathlib")

except Exception as e:
    print(f"Erreur: {e}")
    raise SystemExit(1)
```

```python
# Tests

try:
    assert mathlib.addition(10, 5) == 15.0
    assert mathlib.subtraction(10, 5) == 5.0
    assert mathlib.multiplication(10, 5) == 50.0
    assert mathlib.division(10, 2) == 5.0
    assert mathlib.power(2, 3) == 8.0
    print(f"10 / 0 = {mathlib.division(10, 0)}") # raise error -> Division by Zero

except RuntimeError as e:
    print(f"Error: {e}")
```
