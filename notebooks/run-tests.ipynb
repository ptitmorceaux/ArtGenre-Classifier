### Import Prerequis

```python
import sys
import os
import subprocess

sys.path.append(os.path.abspath(os.path.join('..')))
```

### Functions

```python
def run_cmd(cmd: str):
    try:
        result = subprocess.run(
            cmd,
            shell=True,
            capture_output=True,
            text=True
        )
        print(result.stdout)

        if result.stderr:
            print(result.stderr)
        
        if result.returncode != 0:
            print(f"Failure with exit code {result.returncode}")
            sys.exit(1)

    except Exception as e:
        print(f"Failure: {e}")
        sys.exit(1)
```

### Make

```python
run_cmd("make -C ../libc clean")
run_cmd("make -C ../libc all")
```

### Tests

```python
print("Pylint Ruff check and fix:")
run_cmd("ruff check ../ --fix")
```

```python
print("Running pytest:")
run_cmd("pytest ../")
```
